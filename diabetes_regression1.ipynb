TRAINING



Split X


```python
from sklearn.model_selection import train_test_split 

# Split into validation and training data
X_train, X_val, y_train, y_val = train_test_split(X, y, test_size=0.1, random_state=13)
X_train.shape, X_val.shape, y_train.shape, y_val.shape

#397 veri satırını eğitim için, 45 veri satırını da test için ayırdık.
# test_size ile oynayarak bunların oranını değiştirebilirim.
# Öyleyse eğitim kısmına ne kadar data eklersem hatam o kadar azalmalı. (Ama azalmadı)
```

Define Model

```python
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import VotingRegressor
from sklearn.inspection import permutation_importance

```

Gradient Boosting Regression

Her bir model, önceki modelin yaptığı hatalardan öğrenir.
Bu model, beklenen değeri tekli tahmin şeklinde değil olasılık dağılımını sonuç olarak elde eder.(bu kısımdan emin değilim)


```python
from sklearn.metrics import r2_score, mean_squared_error
```

```python
model1 = GradientBoostingRegressor(random_state=1, n_estimators=500, max_depth=4).fit(X_train, y_train)
model1.fit(X_train, y_train)
print(model1.score(X_train, y_train))

# score() bize modelimizin doğruluğu hakkında bilgi verir. Ama score fonksiyonunu kullanırken, y_pred ve y_val değerlerini fonksiyona girmemiz gerekirken neden X_train, y_train 'i kullanırız.
```

```python
from sklearn.metrics import mean_squared_error

rms = mean_squared_error(y_val, model1.predict(X_val), squared=True)
rms
r2=r2_score(y_val,model1.predict(X_val))
rms,r2
```

```python
feature_importance = model1.feature_importances_
sorted_idx = np.argsort(feature_importance)
pos = np.arange(sorted_idx.shape[0]) + 0.5
fig = plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
plt.barh(pos, feature_importance[sorted_idx], align="center")
plt.yticks(pos, np.array(df.columns)[sorted_idx])
plt.title("Feature Importance (MDI)")

result = permutation_importance(
    model1, X_val, y_val, n_repeats=10, random_state=42, n_jobs=2
)
sorted_idx = result.importances_mean.argsort()
plt.subplot(1, 2, 2)
plt.boxplot(
    result.importances[sorted_idx].T,
    vert=False,
    labels=np.array(df.columns)[sorted_idx],
)
plt.title("Permutation Importance (test set)")
fig.tight_layout()
plt.show()
```

```python
y_val
```

```python
plt.figure(figsize=(5, 5))
plt.scatter(y_val,model1.predict(X_val) , c='crimson')
plt.yscale('log')
plt.xscale('log')

p1 = max(max(model1.predict(X_val)), max(y_val))
p2 = min(min(model1.predict(X_val)), min(y_val))
plt.plot([p1, p2], [p1, p2], 'b-')
plt.xlabel('Actual Values', fontsize=15)
plt.ylabel('Predictions', fontsize=15)
plt.axis('equal')
plt.show()
```

Voting Regression

```python
#Voting Regression, diğer regresyon modellerinin ortalaması olan tahminde bulunur.


model_vote = VotingRegressor([('gb', model1), ('rf', model2), ('lr', model3)],weights=[1,0,1]).fit(X_train, y_train)
print(model_vote.score(X_train, y_train))
```

```python
pred_vote = model_vote.predict(X_val)
pred_vote
```

```python
rms = mean_squared_error(y_val, pred_vote, squared=True)
rms
```

```python
plt.figure(figsize=(5,5))
plt.scatter(y_val, pred_vote, c='crimson')
plt.yscale('log')
plt.xscale('log')

p1 = max(max(pred_vote), max(y_val))
p2 = min(min(pred_vote), min(y_val))
plt.plot([p1, p2], [p1, p2], 'b-')
plt.xlabel('Actual Values', fontsize=15)
plt.ylabel('Predictions', fontsize=15)
plt.axis('equal')
plt.show()
```
